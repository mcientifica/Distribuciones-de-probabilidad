11. Realizar una simulación para estimar la facturación diaria de un local. Se tienen los siguientes datos:
    * El día laboral tiene 8 horas
    * Los clientes arriban según una distribución exponencial de 30 clientes por hora
    * El gasto de cada cliente en el local sigue una distribución normal de media 50 euros y desvío 10 euros

```python
# 8 horas
# lam = 30
# x~N(50,10)
acum=[]
for i in range(10000):
    clientes_por_dia = np.random.poisson(lam = 30, size=8)
    gasto_clientes = np.random.normal(loc = 50, scale = 10,size = sum(clientes_por_dia))
    facturado = sum(gasto_clientes)
    acum.append(facturado)
```

```python
round(sum(acum)/len(acum), 2)
```

## Calcular Probabilidades

### Alturas

1.  Graficar una distribución que represente las alturas de una población, la altura promedio es de 1.7 metros y el desvío estandar es de 0.1 metros.

```python
numeros = np.random.normal(loc = 170, scale = 10, size = 10000)
sns.histplot(numeros)
plt.show()
```

2. Calcular probabilidad de que una persona mida más de 2 metros

```python
personas_altas = len(numeros[numeros > 200])
personas_altas / len(numeros) * 100
```

2. Calcular probabilidad de que una persona mida menos de 1.4 metros

```python
personas_bajas = len(numeros[numeros < 140])
personas_bajas / len(numeros) * 100
```

3. Calcular probabilidad de que una persona mida entre 1.5 metros y 1.7 metros

```python
personas_intermedias = len(numeros[(numeros >= 150) & (numeros <= 170)])
personas_intermedias / len(numeros) * 100
```

4. Calcular probabilidad de que una persona mida exactamente 1.7 metros

```python
persona_170 = len(numeros[numeros == 170])
persona_170 / len(numeros) * 100
```

5. Calcular probabilidad de que una persona no mida entre 1.4 metros y 2 metros

```python
personas_extremos = len(numeros[~((numeros > 140)&(numeros < 200))])
personas_extremos / len(numeros) * 100
```

### Call center

1.  Graficar una distribución que represente las llamadas realizadas por hora a un call center, el cual recibe 8 llamadas por hora

```python
numeros = np.random.poisson(8, 100000)
sns.countplot(x=numeros, color = "red")
plt.show()
```

2. Calcular probabilidad de que llamen entre 9 y 12 personas

```python
llamadas_9_12 = len(numeros[(numeros >=9)&(numeros <= 12)])
llamadas_9_12 / len(numeros) * 100
```

3. Calcular probabilidad de que llamen menos de 2 personas

```python
menos_dos_personas = len(numeros[numeros <= 2])
menos_dos_personas / len(numeros) * 100
```

4. Calcular probabilidad de que llamen exactamente 8 personas

```python
llamadas_ocho = len(numeros[numeros == 8])
llamadas_ocho / len(numeros) * 100
```

5. Calcular la probabilidad de que llamen más de 9 personas

```python
mas_de_nueve = len(numeros[numeros >= 9])
mas_de_nueve / len(numeros) * 100
```

### Ruleta

1. Graficar distribución de la probabilidad que muestre la cantidad de veces que voy a sacar rojo en la ruleta (ganar) al girarla 20 veces. Recordar que hay 18 números rojos y 37 números 

```python
n = 20
p = 18/37

numeros=np.random.binomial(n, p, 100000)
sns.countplot(x=numeros, color="forestgreen")
plt.show()
```

2. Calcular probabilidad de haber ganado 20 veces

```python
ganar_20_veces = len(numeros[numeros == 20])
ganar_20_veces / len(numeros) * 100
```

3. Calcular la probabilidad de ganar más de la mitad de las veces

```python
ganar_mas_mitad = len(numeros[numeros > 10])
ganar_mas_mitad / len(numeros) * 100
```

4. Calcular la probabilidad de haber perdido todas las rondas 

```python
perder_todas = len(numeros[numeros == 0]) 
perder_todas / len(numeros) * 100
```

5. Calcular la probabilidad de haber ganado un número par de veces (2, 4, 6, ..., 20)
